***Filtramos el DataFrame, quedandonos con los registros relacionados al comercio y al estado de nuestro interes; McDonald's en Florida***

Primero filtramos las filas que contengan el nombre McDonald's

```python
df = df[df['name'].str.contains(r"\bMcDonald's\b|\bMc Donald's\b", case=False, na=False, regex=True)]
```

Ahora filtramos los McDdonald's ubicados en Florida mediante el zip code 32003 y 34997

Fuente https://codigo-postal.co/eeuu/florida/

```python
df = df[df['address'].str.extract(r"\b(32[0-9]{3}|33[0-9]{3}|34[0-9]{3}|349[0-7]{1})\b", expand=False).notnull()].copy()
```

Verificamos el nombre de los datos filtrados

```python
df['name'].unique()
```

Se observa que hay varios nombres, los miramos a mas detalles para saber si son comida rapida

```python
nombres = ["Martin Brower/McDonald's Distribution Center","McDonald's - NEW LOCATION (next to Walgreen's) NOW OPEN","McDonald's / Ross's"]
nombres_filtrado = df[df['name'].isin(nombres)]
nombres_filtrado
```

Observamos que solo un local de los tres revisados tienen como categoría "Fast food". Procederemos a eliminar los otros que no pertecen a esta categoría

```python
nombres = ["Martin Brower/McDonald's Distribution Center","McDonald's / Ross's"]
df = df[~df['name'].isin(nombres)]
```

Verificamos los tipos de datos y valores nulos

```python
ut.tipo_dato(df)
```

La columna category no tiene el formato adecuado, hemos creado una funcion para normalizar la columna

```python
df['category'].unique()
```

```python
df = ut.categoria_metadatos(df)
df.head(2)
```

Eliminamos columnas que no son relevantes

```python
df = df.drop(columns=['Unnamed: 0','relative_results','price','MISC','hours'])
```

La columna description presenta dos tipos de clases ya que posee nulos

```python
# Seleccionamos los valores nulos
valores_nulos = df[df['description'].isnull()]
valores_nulos
```

Reemplazamos valores nulos existentes

```python
df[['description','category','state']] = df[['description','category','state']].fillna('Sin datos')
# Filtro por el valor 'Sin datos' para mostrar el resultado
df[df['description'] == 'Sin datos']
```

Renombramos la columna state, que hace referencia a la hora de apertura y cierre del local

```python
df = df.rename(columns={'state':'open_closes'})
```

```python
df.head(2)
```

La columna **address**, ademas de contener la dirección del comercio contiene el nombre del mismo, la ciudad, el estado, etc. A traves de la función **extract_address**, extraeremos la dirección, sin todo lo demas:

```python
df['address'] = df['address'].apply(ut.extract_address)
```

Normalizamos los valores de la columna **address**, colocando la primer letra dencada palabra en mayuscula y el resto en minuscula:

```python
df['address'] = df['address'].apply(lambda x: x.title())
```
