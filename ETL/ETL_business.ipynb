***Filtramos el DataFrame, quedandonos con los registros relacionados al comercio y al estado de nuestro interes; McDonald's en Florida***

Primero filtramos las filas que contengan el nombre McDonald

```python
df_business_mc = df_business[df_business['name'].str.contains("McDonald")]
df_business_mc.head(3)
```

Para realizar el filtro a los negocios Ãºbicados en Florida, filtraremos por postal_code y por city.

Primero filtraremos utilizando el codigo postal de Florida. Para esto, primero convertimos la columna postal_code a numerico

```python
df_business_mc.loc[:, 'postal_code'] = pd.to_numeric(df_business_mc['postal_code'], errors='coerce')
```

Ahora filtramos las filas por codigos postales entre 32003 y 34997

Fuente https://codigo-postal.co/eeuu/florida/

```python
df_business_mc_fl = df_business_mc[(df_business_mc['postal_code'] >= 32003) & (df_business_mc['postal_code'] <= 34997)].copy()
```

Ahora filtraremos las filas por ciudades. Para esto, utilizaremos un dataset de ciudades de Florida

Fuente https://worldpopulationreview.com/states/cities/florida

```python
df_ciudades_fl = pd.read_csv('../Datos/Archivos csv/ciudades-florida.csv')
df_ciudades_fl.head(2)
```

Almacenamos los nombres de las ciudades en una lista

```python
nombres_ciudades = df_ciudades_fl['name'].tolist()
```

Filtramos el dataframe quedandonos solo con las filas donde aparezcan los nombres que contiene la lista anteriormente creada

```python
df_business = df_business_mc_fl[df_business_mc_fl['city'].isin(nombres_ciudades)]
df_business.head(3)
```

Corregimos los acrÃ³nimos de la columna state, ya que las filas han sido debidamente filtradas y aÃºn asÃ­ presenta acrÃ³nimos erroneos

```python
df_business.loc[:, 'state'] = 'FL'
df_business.head(2)
```

Eliminaremos la columna 'categories' ya que es irrelevante pues ya sabemos que estamos trabajando con comercios de comida rÃ¡pida.

```python
columna = ['categories']
df_business = df_business.drop(columna, axis=1)
df_business.head(2)
```

Verificamos el tipo de datos de las columnas

```python
ut.tipo_dato(df_business)
```

Reseteamos el indice del DataFrame

```python
df_business.reset_index(drop = True, inplace = True)
```

Desanidamos la columna hours, contiene la hora de apertura y cierre del local todos los dÃ­as de la semana

```python
# Aplicamos la funciÃ³n para desanidar y creamos las nuevas columnas
df_business['hours'] = df_business['hours'].apply(ut.horas_business)

# Concatenamos las nuevas columnas al DataFrame original
df_business = pd.concat([df_business, df_business['hours'].apply(pd.Series)], axis=1)

# Eliminamos la columna 'hours'
df_business = df_business.drop(['hours'], axis = 1)
```
